### **Step 1: Creating a Simple Undirected Graph**

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an undirected graph
G = nx.Graph()

# Add nodes
G.add_nodes_from([1, 2, 3, 4, 5, 6])

# Add edges
G.add_edges_from([(1, 2), (1, 3), (3, 4), (4, 5), (2, 5), (1, 6), (2, 6)])

# Draw the graph
plt.figure(figsize=(6, 6))
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=1500, font_size=12)
plt.title("Undirected Graph")
plt.show()
```

### **Step 2: Generating Directed Graphs**
A **Directed Graph (Digraph)** is a graph where each edge has a direction.

```python
DG = nx.DiGraph()
DG.add_edges_from([(1, 2), (2, 3), (3, 1), (3, 4)])

plt.figure(figsize=(6, 6))
nx.draw(DG, with_labels=True, node_color="lightcoral", edge_color="black", node_size=1500, arrows=True)
plt.title("Directed Graph")
plt.show()
```

### **Step 3: Generating Weighted Graph**
A **Weighted Graph** assigns numerical values to edges, often representing distance or cost.

```python
WG = nx.Graph()
WG.add_weighted_edges_from([(1, 2, 3.5), (2, 3, 2.0), (3, 4, 1.2), (4, 1, 4.8)])

plt.figure(figsize=(6, 6))
pos = nx.spring_layout(WG)
nx.draw(WG, pos, with_labels=True, node_color="lightgreen", edge_color="black", node_size=1500)

# Add edge labels (weights)
edge_labels = {(u, v): f"{d['weight']}" for u, v, d in WG.edges(data=True)}
nx.draw_networkx_edge_labels(WG, pos, edge_labels=edge_labels)

plt.title("Weighted Graph")
plt.show()
```

### **Step 4: Creating a MultiGraph (Multiple Edges)**

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Create a MultiGraph with multiple edges
G = nx.MultiGraph()
edges = [(1, 2), (1, 2), (2, 3), (3, 4), (2, 3), (3, 4), (4, 1)]
G.add_edges_from(edges)

# Define the layout
pos = nx.spring_layout(G, seed=42)  # Use spring layout for positioning

# Plot the nodes
plt.figure(figsize=(6, 6))
nx.draw_networkx_nodes(G, pos, node_color="lightgreen", node_size=700)

# Draw curved edges using Bezier curves
ax = plt.gca()
for edge in G.edges(keys=True):
    u, v, key = edge
    rad = 0.3 * (key + 1)  # Adjust curvature for multiple edges
    nx.draw_networkx_edges(G, pos, edgelist=[(u, v)], connectionstyle=f'arc3, rad={rad}', edge_color="black")

# Draw labels
nx.draw_networkx_labels(G, pos, font_size=12, font_color="black")

# Title and display
plt.title("MultiGraph with Curved Multiple Edges")
plt.axis("off")
plt.show()
```

### **Step 6: Random Graph (Erdős-Rényi Model)**

The **Erdős-Rényi (ER) model** is one of the **fundamental random graph models** used in network science and graph theory. It was introduced by **Paul Erdős and Alfréd Rényi** in the 1950s and serves as a foundational framework for understanding **random networks**. There are two main variants of the Erdős-Rényi model:

1. **$ G(n, p)$ Model (Probability-based).** In this setting, a graph is constructed with $ n $ nodes. Each possible edge **between any two nodes** is included with **independent probability** $ p $, where $ 0 \leq p \leq 1 $. The probability $ p $ controls the **density** of the graph:  
     - If $ p = 0 $, the graph has **no edges** (empty graph).  
     - If $ p = 1 $, the graph is a **complete graph** where every node connects to every other node.  

2. **$ G(n, m) $ Model (Edge-based).** In this setting on the other hand, a graph is constructed with $ n $ nodes and exactly **$ m $ edges** chosen uniformly at random from all possible edges. Unlike $ G(n, p) $, this model **fixes** the number of edges instead of defining a probability for each edge.  

#### **Properties of Erdős-Rényi Graphs**  
- The degree distribution follows a **binomial distribution**:  
  $$
  p(k) = \binom{n-1}{k} p^k (1 - p)^{(n-1-k)}
  $$
  which approximates a **Poisson distribution** for large $ n $.  
- Erdős-Rényi graphs **do not exhibit scale-free behavior** (unlike Barabási–Albert networks).  
- There is a **threshold phenomenon**:  
  - If $ p < \frac{\ln n}{n} $, the graph is **disconnected** (many isolated components).  
  - If $ p > \frac{\ln n}{n} $, the graph becomes **connected** with high probability.  

Here is a Python program that **simulates and visualizes** both main variants of the **Erdős-Rényi models**:  
- **$ G(n, p) $ model** (where each edge appears with independent probability $ p $)  
- **$ G(n, m) $ model** (where exactly $ m $ edges are chosen at random)oges is fixed at \( m \).

```python
import networkx as nx
import matplotlib.pyplot as plt

# Set parameters
n = 20  # Number of nodes
p = 0.3  # Probability of an edge in G(n, p)
m = 30   # Number of edges in G(n, m)

# Generate Erdős-Rényi graphs
G_np = nx.erdos_renyi_graph(n=n, p=p)  # G(n, p) model
G_nm = nx.gnm_random_graph(n=n, m=m)   # G(n, m) model

# Define layout for consistent positioning
pos = nx.spring_layout(G_np, seed=42)  # Use the same layout for fair comparison

# Plot side-by-side visualizations
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot G(n, p) graph
axes[0].set_title(f"Erdős-Rényi Graph G({n}, {p})")
nx.draw(G_np, pos, ax=axes[0], with_labels=True, node_color="lightblue", edge_color="gray", node_size=600, font_size=8)

# Plot G(n, m) graph
axes[1].set_title(f"Erdős-Rényi Graph G({n}, {m})")
nx.draw(G_nm, pos, ax=axes[1], with_labels=True, node_color="lightcoral", edge_color="black", node_size=600, font_size=8)

# Adjust layout and save
plt.tight_layout()
plt.savefig("erdos_renyi_graphs.png", dpi=600)
plt.show()
```

## **Step 7. Creating a Barabási–Albert Scale-Free Network**

A **Barabási–Albert (BA) Scale-Free Network** is a type of **random network model** that follows the **preferential attachment** mechanism, meaning that **new nodes prefer to connect to already well-connected nodes** (i.e., hubs). This process leads to a **power-law degree distribution**, where a small number of nodes have many connections, while most nodes have only a few. The **Barabási–Albert model** is defined by the following **growth and preferential attachment** rules:

1. **Growth.** Start with a small connected network of $m_0$ nodes. At each step, add a new node with $m$ edges.

2. **Preferential Attachment.** The probability $p(i)$ of a new node connecting to an existing node $i$ is proportional to its degree $d_i$:

   $$
   p(i) = \frac{d_i}{\sum_j d_j}
   $$

   where $d_i$ is the degree of node $i$, and the denominator sums the degrees of all nodes in the network.

### **Key Properties of Barabási–Albert Networks**
✅ **Scale-Free Nature.** The degree distribution follows a **power-law**:

   $$
   p(d) \sim d^{-\gamma}, \quad \text{where } \gamma \approx 3
   $$

✅ **Emergence of Hubs.** A few nodes (hubs) dominate the network with high connectivity.  

✅ **Robustness & Fragility:**  
   - **Robust against random failures** (most nodes are low-degree).  
   - **Vulnerable to targeted attacks** (removing hubs can break the network).  

✅ **Common in Real-World Networks.** Found in the **internet**, **social networks**, **citation networks**, and **biological systems** (e.g., protein interactions).  

Below is a **Python program** using **NetworkX** to generate and visualize a **BA scale-free network**:this with **degree distribution plots** or **further analysis**? 🚀

```python
import networkx as nx
import matplotlib.pyplot as plt

# Generate a Barabási–Albert scale-free network with 100 nodes, each new node connecting to 2 existing nodes
n = 100  # Total number of nodes
m = 2    # Each new node connects to 'm' existing nodes
G = nx.barabasi_albert_graph(n, m)

# Define layout for visualization
pos = nx.spring_layout(G, seed=42)

# Plot the network
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=False, node_size=50, node_color="orange", edge_color="gray", alpha=0.7)

# Add title and display
plt.title("Barabási–Albert Scale-Free Network", fontsize=14)
plt.show()
```

### **Step 11. Creating a Balanced Tree**
In graph theory, the **height of a tree** is defined as the **length of the longest path** from the root node to any of its leaf nodes. It is the maximum number of edges in the longest path from the root to a leaf.

Mathematically, if $ T $ is a tree with root $ r $, then the height of the tree $ h(T) $ is:
$$
h(T) = \max_{v \in \text{leaves}} d(r, v)
$$
where $ d(r, v) $ represents the **shortest path length** (number of edges) from the root $ r $ to a leaf node $ v $.


A **balanced tree** is a tree data structure in which the height (or depth) is kept as small as possible relative to the number of nodes. Balanced trees are crucial in **graph theory** and **computer science**, particularly in search algorithms, database indexing, and network routing.

A tree $ T $ is considered **balanced** if it satisfies one of the following conditions:
1. **Height-Balanced (AVL Trees)**  
   A tree is height-balanced if, for every node $ v $, the heights of its left and right subtrees differ by at most 1:
   $$
   |h_{\text{left}} - h_{\text{right}}| \leq 1
   $$
   where $ h_{\text{left}} $ and $ h_{\text{right}} $ are the heights of the left and right subtrees of $ v $.

2. **Weight-Balanced**  
   A tree is weight-balanced if the number of nodes in each subtree is proportional to the total number of nodes.  
   This ensures that the search time remains logarithmic.

```python
G = nx.balanced_tree(2, 3)  # Binary tree of height 3

plt.figure(figsize=(5, 5))
nx.draw(G, with_labels=True, node_color="brown", edge_color="black", node_size=500)
plt.title("Balanced Tree (2,3)")
plt.show()
```

# **Plotting Graphs Using the Adjacency Matrix in NetworkX**

The goal of this activity is to understand how to represent graphs using an **Adjacency Matrix** and visualize them using the **NetworkX** library.

## **Task Overview**

### **Step 1: Import Necessary Libraries**

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

## **Step 2: Define an Adjacency Matrix**
An **Adjacency Matrix** is a square matrix used to represent a finite graph. The elements of the matrix indicate whether pairs of vertices are adjacent or not.

```python
# Define an adjacency matrix for an undirected graph
adj_matrix = np.array([
    [0, 1, 1, 0, 1],  # Node 0 is connected to Nodes 1, 2 and 4
    [1, 0, 1, 1, 0],  # Node 1 is connected to Nodes 0, 2, and 3
    [1, 1, 0, 1, 1],  # Node 2 is connected to Nodes 0, 1, 3, and 4
    [0, 1, 1, 0, 1],  # Node 3 is connected to Nodes 1, 2, and 4
    [0, 0, 1, 1, 0]   # Node 4 is connected to Nodes 0, 2 and 3
])
```

### **Step 3: Create a Graph from the Adjacency Matrix**
Convert the matrix into a **NetworkX** graph.


```python
# Use `from_numpy_array` instead of the deprecated `from_numpy_matrix`
G = nx.from_numpy_array(adj_matrix)

# Define different layouts for visualization
layouts = [nx.spring_layout, nx.circular_layout, nx.shell_layout]

# Plot the graph using different layouts
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, layout in enumerate(layouts):
    pos = layout(G)  # Get the layout position
    axes[i].set_title(layout.__name__)  # Set title as layout name
    nx.draw(G, pos, with_labels=True, node_size=700, node_color="lightblue", edge_color="black", font_size=10, ax=axes[i])

# Display the plots
plt.tight_layout()
plt.show()
```

### **Step 4: Visualize the Graph**
Use **NetworkX** layouts to plot the graph.

```python
# Define layouts for visualization
layouts = [nx.spring_layout, nx.circular_layout, nx.shell_layout]

# Plot the graph using different layouts
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, layout in enumerate(layouts):
    pos = layout(G)  # Get the layout position
    axes[i].set_title(layout.__name__)  # Set title as layout name
    nx.draw(G, pos, with_labels=True, node_size=700, node_color="lightblue", edge_color="black", font_size=10, ax=axes[i])

# Display the plots
plt.tight_layout()
plt.show()
```
